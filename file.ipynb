```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,precision_score
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
```

```python
data = load_iris()
data['data']
new_data = pd.DataFrame(data['data'],columns=['1','2','3','4'])
# new_data = pd.DataFrame(data['data'],columns=data)
new_data['target'] = data['target']
print(new_data)

x = new_data[['1','2','3','4']]
y = new_data['target']

scaler = StandardScaler()
scaler.fit_transform(x)

# clf = RandomForestClassifier()
clf = KNeighborsClassifier()

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

clf.fit(x_train,y_train)

y_pred = clf.predict(x_test)

p_score = precision_score(y_pred,y_test,average='weighted')

print(p_score)
```

```python
import matplotlib.pyplot as plt

def scatter_plot_matplotlib(x, y):
    plt.figure(figsize=(8,6))
    plt.scatter(x['1'], x['3'], c=y, cmap='viridis', edgecolor='k', s=100)
    plt.title('Scatter plot using Matplotlib', fontsize=14)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 3')
    plt.colorbar(label='Target')
    plt.show()

# Usage
scatter_plot_matplotlib(x, y)
```

```python
import plotly.express as px

def scatter_plot_plotly(x, y):
    fig = px.scatter(x=x['1'], y=x['2'], color=y, labels={'x': 'Feature 1', 'y': 'Feature 2'}, 
                     title="Scatter plot using Plotly Express")
    fig.update_traces(marker=dict(size=10, line=dict(width=2, color='DarkSlateGrey')))
    fig.show()

# Usage
scatter_plot_plotly(x, y)
```
